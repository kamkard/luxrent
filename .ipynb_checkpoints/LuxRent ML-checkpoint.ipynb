```python
import sqlite3
from sqlite3 import Error
import datetime
import pandas as pd
import numpy as np
from IPython.display import display, HTML
from sqlalchemy import create_engine
from jupyterthemes import jtplot

pd.options.display.max_rows = 20
pd.options.display.float_format = '{:.1f}'.format

#Panda's method of usiing SQLalchemy to grab my SQLite data
engine = create_engine('sqlite:///C:/Users/TheDave/Desktop/LuxData.sqlite')

#Initialize the whole Lux Data
amli_austin_df = pd.read_sql_query("select * from Lux", engine)
#Drop ID column
amli_austin_df = amli_austin_df.drop('id',1)
amli_austin_df.describe(include='all')
amli_austin_df.head()
```

Input = Building, Floorplan, Floor, Price 
Target = Regression for how long open
Classification if rented by next month (Y or N)
Synth Data = Address ,Price changed? (Need to check these data points)

```python
#Need to make targets based on if rented next month

#amli_austin_df.Date.unique()
#Date Range = 28/10/17 - 24/02/18 . 118 rows would be max

amli_austin_df.Roomnumber.nunique()
# 579 unique rooms from all buildings.  
#Duplicate room numbers from different buildings missing

buildings = amli_austin_df.Building.unique()
uRooms = []
uRoomsTot = []
buildS = []

for b in buildings:
    u = amli_austin_df[amli_austin_df['Building'] == b].Roomnumber.nunique()
    uRooms.append(u)
    
    uR = amli_austin_df[amli_austin_df['Building'] == b].Roomnumber.unique()
    uRoomsTot.extend(uR)
    
    
    buildS.extend([b]*u)
    
uRooms = pd.Series(uRooms)
uRooms.sum()

uRoomsTot = pd.Series(uRoomsTot)
#590 Rooms total, meaning earlier method was losing some!!!  
#Have list by building in order not to lose all rooms

#uniqueRooms = amli_austin_df.Roomnumber.unique()
#uRoomsTot

buildRoomsTot = pd.DataFrame({'Building' : buildS, 'Roomnumber' : uRoomsTot})

#Have BuildRoomsTot with all unique rooms from each building.  
#Need to find how long rooms were open for

delTime = []
buildName = []
fPlan = []
fNum = []
roomPrice = []


#buildRoomsTot.iloc[3,1]
#amli_austin_df[amli_austin_df['Building'] == buildRoomsTot.iloc[bName,0]]
for i in range(len(buildRoomsTot)):
    amliURooms = amli_austin_df[
        (amli_austin_df['Building'] == buildRoomsTot.iloc[i,0]) & 
        (amli_austin_df['Roomnumber'] == buildRoomsTot.iloc[i,1])]

    #Features
    fPlan.append(amliURooms.Floorplan.iloc[0])
    fNum.append(amliURooms.Floor.iloc[0])
    roomPrice.append(amliURooms.Price.iloc[0])
    
    #Target
    fTime = amliURooms.Date.iloc[0]
    lTime = amliURooms.Date.iloc[-1]
    lTime = datetime.datetime.strptime(lTime,"%d/%m/%y")
    fTime = datetime.datetime.strptime(fTime,"%d/%m/%y")
    
    delTime.append(abs(lTime - fTime).days)

#RoomFeatures is Features

fPlan = pd.Series(fPlan)

fNum = pd.Series(fNum)
fNum = pd.to_numeric(fNum,errors='coerce').fillna(0).astype(np.int64)

roomPrice = pd.Series(roomPrice)
roomPrice=roomPrice.str.replace('$','')
roomPrice= roomPrice.str.replace(',','')
roomPrice=pd.to_numeric(roomPrice,errors='coerce')

buildRoomsTot["Floor Plan"] = fPlan.values
buildRoomsTot["Floor Number"] = fNum.values
buildRoomsTot["Price"] = roomPrice.values

buildRoomsTot["Binned Price"] = pd.cut(buildRoomsTot["Price"],10) 

roomFeatures = buildRoomsTot[['Floor Number','Binned Price']]
roomFeatures = pd.concat([roomFeatures,pd.get_dummies(buildRoomsTot.Building)],axis=1)

#DelTime is Target
delTime = pd.Series(delTime)

#buildRoomsTot is total dataframe
buildRoomsTot["Days Open"] = delTime.values
#buildRoomsTot.describe(include='all')
roomFeatures.describe(include='all')
#daysOpen = abs(lTime - fTime)
#daysOpen.days
#amli_austin_df[amli_austin_df['Roomnumber'] == uniqueRooms[292]]
#amli_austin_df[amli_austin_df['Roomnumber'] == uniqueRooms[6]].iloc[0]
#amli_austin_df[amli_austin_df['Roomnumber'] == uniqueRooms[6]].tail(1)

```
